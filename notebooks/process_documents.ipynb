```python
import os
import sys
import json
from pathlib import Path
if "../src" not in sys.path:
    sys.path.append("../src")
import utils
import settings
from pdf_processor import PdfProcessor

wrapper = utils.get_text_wrapper()
```

```python
directories = os.listdir(settings.DATA_DIR)

for dir in directories:
    dir_path = os.path.join(settings.DATA_DIR, dir)

    files = [file for file in os.listdir(dir_path) if os.path.isfile(os.path.join(dir_path, file))]
    files.remove('metadata.json')

    metadata_filepath = os.path.join(dir_path, 'metadata.json')
    with open(metadata_filepath, "r") as file:
        metadata = json.load(file)

    for item in metadata:
        if item["path"] not in files:
            continue
        
        file_extension = Path(item["path"]).suffix
        filepath = os.path.join(dir_path, item["path"])
        output_filepath = filepath.split('.')[0] + ".json"


        if os.path.isfile(output_filepath):
            print(f"File {output_filepath} already exists. Skipping file...")
            continue

        if file_extension == ".pdf":
            documentProcessor = PdfProcessor(filepath, item["url"])

        elif file_extension == ".txt":
            # documentProcessor = TxtProcessor(filepath, item["url"])
            continue
        
        print(f"Processing file {filepath}...")
        
        documentProcessor.process()
        documentProcessor.export_chunked_document()
```
